```python
import pandas as pd
import numpy as np
import datetime
```

```python
def loadturndata(date):
    strdate = str(date)
    filename = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_'+strdate+'.txt'
    df = pd.read_csv(filename)
    #cleanup/standardize coulmn names
    df = df.rename(columns=lambda x: x.strip().lower())
    #combine date and time columns
    df['datetime'] = df['date'] + ' ' + df['time']
    #from object to datetime
    df['datetime_clean'] = [datetime.datetime.strptime(x, '%m/%d/%Y %H:%M:%S') for x in df['datetime']]
    return df
```

```python
dates = datelist()
```

```python
def loadturnlist(dates):
    data = pd.DataFrame()
    x = []
    for i in dates:
        df = (loadturndata(i))
        x.append(df)
    data = pd.concat(x)
    return data
    
```

```python
final = loadturnlist(dates)
```
